```python
adreno2 = adreno.loc[:,['Patient ID','Diagnosis Age','Aneuploidy Score',
            'Cancer Type','Last Communication Contact from Initial Pathologic Diagnosis Date',
            'Birth from Initial Pathologic Diagnosis Date','Disease Free (Months)',
            'Disease Free Status','Months of disease-specific survival','Disease-specific Survival status',
            'Ethnicity Category','Form completion date','Fraction Genome Altered','Genetic Ancestry Label',
            'MSI MANTIS Score','MSIsensor Score','Mutation Count','New Neoplasm Event Post Initial Therapy Indicator',
            'Overall Survival (Months)','Overall Survival Status','Person Neoplasm Cancer Status',
            'Progress Free Survival (Months)','Progression Free Status','Primary Lymph Node Presentation Assessment',
            'Prior Diagnosis','Race Category','Radiation Therapy','Sex',
            'Tumor Break Load','Tissue Prospective Collection Indicator','Tissue Retrospective Collection Indicator',
            'TMB (nonsynonymous)']]
```

```python
adreno2
```

```python
adreno2['Disease_Free_Months'] = round(adreno2['Disease Free (Months)'])
adreno2['Months of Survival'] = round(adreno2['Months of disease-specific survival'])
adreno2['Fraction_Genome_Altered'] = round(adreno2['Fraction Genome Altered'],2)
adreno2['MSI_MANTIS_Score'] = round(adreno2['MSI MANTIS Score'],2)
adreno2['Overall_Survival_Months'] = round(adreno2['Overall Survival (Months)'])
adreno2['Progress_Free_Survival_Months'] = round(adreno2['Progress Free Survival (Months)'])
adreno2['TMB'] = round(adreno2['TMB (nonsynonymous)'],2)
```

```python
adreno2
```

```python
adreno2['Percent_Genome_Altered'] = round(adreno2['Fraction Genome Altered']*100,2)
```

```python
def race_var(row):
    if row['Race Category'] == "White" and row['Ethnicity Category'] == "Hispanic Or Latino":
        return "Hispanic"
    elif row['Race Category'] == "Black or African American":
        return "Black"
    else:
        return row['Race Category']


adreno2['Race'] = adreno2.apply(race_var,axis=1)
```

```python
adreno2
```

```python
adreno2['Last Contact From Initial Diagnosis Date in Months'] = round(adreno2['Last Communication Contact from Initial Pathologic Diagnosis Date']/30)
```

```python
adreno2
```

```python
def survival_var(row):
    if row['Disease-specific Survival status'] == "0:ALIVE OR DEAD TUMOR FREE" and row['Overall Survival Status'] == "1:DECEASED":
        return "Dead w/o Tumor"
    elif row['Disease-specific Survival status'] == "0:ALIVE OR DEAD TUMOR FREE" and row['Person Neoplasm Cancer Status'] == "With Tumor" and row['New Neoplasm Event Post Initial Therapy Indicator'] == "No":
        return "Unknown"
    elif row['Disease-specific Survival status'] == "0:ALIVE OR DEAD TUMOR FREE" and row['Person Neoplasm Cancer Status'] == "With Tumor" and row['New Neoplasm Event Post Initial Therapy Indicator'] == "Yes":
        return "Alive w/ Tumor"
    elif row['Disease-specific Survival status'] == "1:DEAD WITH TUMOR":
        return "Dead w/ Tumor"
    else:
        return "Alive Tumor-Free"


adreno2['Survival Status'] = adreno2.apply(survival_var,axis=1)
```

```python
adreno2
```

```python
adreno2['Patient_ID'] = adreno2['Patient ID'].str[5:]
```

```python
adreno2
```

```python
def progress_var(row):
    if row["New Neoplasm Event Post Initial Therapy Indicator"] == "Yes":
        return "Recurrence"
    if row['Progression Free Status'] == "1:PROGRESSION":
        return "Progression"
    else:
        return "Neither"

adreno2['Tumor Progression or Recurrence'] = adreno2.apply(progress_var, axis=1)
```
