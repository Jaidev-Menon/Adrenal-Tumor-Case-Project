```python
pheo2['Disease_Free_Months'] = round(pheo2['Disease Free (Months)'])
pheo2['Months of Survival'] = round(pheo2['Months of disease-specific survival'])
pheo2['Fraction_Genome_Altered'] = round(pheo2['Fraction Genome Altered'],2)
pheo2['MSI_MANTIS_Score'] = round(pheo2['MSI MANTIS Score'],2)
pheo2['Overall_Survival_Months'] = round(pheo2['Overall Survival (Months)'])
pheo2['Progress_Free_Survival_Months'] = round(pheo2['Progress Free Survival (Months)'])
pheo2['TMB'] = round(pheo2['TMB (nonsynonymous)'],2)
```

```python
pheo2['Percent_Genome_Altered'] = round(pheo2['Fraction Genome Altered']*100,2)
```

```python
def race_var(row):
    if row['Race Category'] == "White" and row['Ethnicity Category'] == "Hispanic Or Latino":
        return "Hispanic"
    elif row['Race Category'] == "Black or African American":
        return "Black"
    else:
        return row['Race Category']


pheo2['Race'] = pheo2.apply(race_var,axis=1)
```

```python
pheo2['Last Contact From Initial Diagnosis Date in Months'] = round(pheo2['Last Communication Contact from Initial Pathologic Diagnosis Date']/30)
```

```python
def survival_var(row):
    if row['Disease-specific Survival status'] == "0:ALIVE OR DEAD TUMOR FREE" and row['Overall Survival Status'] == "1:DECEASED":
        return "Dead w/o Tumor"
    elif row['Disease-specific Survival status'] == "0:ALIVE OR DEAD TUMOR FREE" and row['Person Neoplasm Cancer Status'] == "With Tumor" and row['New Neoplasm Event Post Initial Therapy Indicator'] == "No":
        return "Unknown"
    elif row['Disease-specific Survival status'] == "0:ALIVE OR DEAD TUMOR FREE" and row['Person Neoplasm Cancer Status'] == "With Tumor" and row['New Neoplasm Event Post Initial Therapy Indicator'] == "Yes":
        return "Alive w/ Tumor"
    elif row['Disease-specific Survival status'] == "1:DEAD WITH TUMOR":
        return "Dead w/ Tumor"
    else:
        return "Alive Tumor-Free"


pheo2['Survival Status'] = pheo2.apply(survival_var,axis=1)
```

```python
def progress_var(row):
    if row["New Neoplasm Event Post Initial Therapy Indicator"] == "Yes":
        return "Recurrence"
    if row['Progression Free Status'] == "1:PROGRESSION":
        return "Progression"
    else:
        return "Neither"

pheo2['Tumor Progression or Recurrence'] = pheo2.apply(progress_var, axis=1)
```
